```python
# eda_02_05 Histogram of Black Rock Forest data
import matplotlib.pyplot as plt

# load data from text fole
D = np.genfromtxt('brf_temp.txt', delimiter='\t')
[N, M]=D.shape;
t = np.zeros((N,1));
d = np.zeros((N,1));
t[:,0] = D[:,0];
d[:,0] = D[:,1];

# histogram
Lh = 100;
dmin = np.min(d);
dmax = np.max(d);
c, e = np.histogram(d,Lh,(dmin,dmax));
Nc = len(c);
Ne = len(e);
count = np.zeros((Nc,1));
count[:,0]=c;
edges = np.zeros((Ne-1,1));
edges[:,0]=e[0:Ne-1];


# plot
fig1 = plt.figure(1,figsize=(8,4));
ax1 = plt.subplot(1,1,1);
plt.axis([-60, 40, 0, np.max(c)]);
plt.plot(edges,count,'k-');
plt.xlabel('temperature (C)');
plt.ylabel('counts');
plt.title('Histogram of Temperature Measurements');
plt.show();
```

```python
# eda_02_07 Histogram of Black Rock Forest data, with density plot

# load data from text file
D = np.genfromtxt('brf_temp.txt', delimiter='\t')
[N, M]=D.shape;
t = np.zeros((N,1));
d = np.zeros((N,1));
t[:,0] = D[:,0];
d[:,0] = D[:,1];

offset = 1000;
Lh = 100;
Lw = floor( N/offset);
Dw = np.zeros((Lh,Lw));

# histograms
dmin = -60;
dmax = 40;
for i in range(Lw):
    j=(i-1)*offset;
    k=j+offset;
    c, e = np.histogram(d[j:k,0],Lh,(dmin,dmax));
    Nc = len(c);
    Dw[:,i]=c;
    

# plot
fig1 = plt.figure(1,figsize=(8,4));
ax1 = plt.subplot(1,1,1);
plt.rcParams['xtick.bottom'] = False;
plt.rcParams['xtick.labelbottom'] = False;
plt.rcParams['xtick.top'] = True;
plt.rcParams['xtick.labeltop'] = True;
plt.xlabel('time (days)');
plt.ylabel('temperature (C)');
ax1.xaxis.set_label_position('top');
plt.axis([0, np.max(t), dmin, dmax]);
plt.gca().invert_yaxis();

# greyscale colormap
bw = np.zeros((256,4));
v = 0.9*(256 - np.linspace( 0, 255, 256 ))/255;
bw[:,0] = v;
bw[:,1] = v;
bw[:,2] = v;
bw[:,3] = np.ones(256);
bwcmap = ListedColormap(bw);

# alternatively, could have ued colormap
# bwcmap = cm.gray;
# bwcmap = bwcmap.reversed();

plt.imshow( Dw, cmap=bwcmap, vmin=0, vmax=np.max(Dw), aspect='auto', extent=(0,np.max(t),dmin,dmax), origin='lower' );
plt.colorbar(label='counts');
plt.show();
```

```python
# eda02_10 finite differnce derivative

# load data from text file
D = np.genfromtxt('neuse.txt', delimiter='\t')
[N, M]=D.shape;

# time in column 1, discharge in colun 2
t = D[:,0];
d = D[:,1];

tmin=500;
tmax = 1000;

# derivative
dddt = np.divide( d[1:N]-d[0:N-1], t[1:N]-t[0:N-1] );

# plot the sample data
fig1 = plt.figure(1,figsize=(12,8));
plt.rcParams['xtick.bottom'] = False;
plt.rcParams['xtick.labelbottom'] = False;
plt.rcParams['xtick.top'] = True;
plt.rcParams['xtick.labeltop'] = True;

# subplot 1 of data
ax1 = plt.subplot(1,3,1);
plt.axis([ 0, 20000, tmin, tmax,]);
plt.plot(d,t,'k-');
ax1.invert_yaxis();
plt.xlabel('time (s)');
plt.ylabel('data d');
plt.title('discharge d');

# subplot 2 of derivative
ax2 = plt.subplot(1,3,2);
plt.axis([-5000, 5000, tmin, tmax, ]);
ax2.invert_yaxis();
plt.plot(dddt,t[0:N-1],'k-');
plt.xlabel('time (s)');
plt.ylabel('dd/dt');
plt.title('derivative dd/dt');

# histogram
Lh = 1000;
dmin = np.min(dddt);
dmax = np.max(dddt);
c, e = np.histogram(dddt,Lh,(dmin,dmax));
Nc = len(c);
Ne = len(e);
count = np.zeros((Nc,1));
count[:,0]=c;
edges = np.zeros((Ne-1,1));
edges[:,0]=e[0:Ne-1];

# subplot 3 of histogram
ax3 = plt.subplot(1, 3, 3);
plt.axis([0, np.max(c), -500, 500]);
ax3.invert_yaxis();
plt.plot(count,edges,'k-');
plt.plot([0,500],[0,0],'k:');
plt.xlabel('counts');
plt.ylabel('dddt');
```
